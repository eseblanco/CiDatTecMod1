## Descripcion del conjunto de datos:

Contexto

El conjunto pertenece a datos registrados por nodos IoT durante los mes de enero a marzo 2022.

Se trata de una serie temporal y contiene en general la siguiente información:

1. Numero de la transacción.
1. Fecha y hora de inicio y de fin de la transacción.
1. Cantidad de litros de combustible despachado.
1. Volumen real registrado por otros nodos IoT en los tanques durante el inicio y fin de la descarga.
1. Estación, sutidor, empresa de origen.
1. Tipo de combustible transaccionado.
1. Geolocalización (si es soportada).
1. Unidad de consumo de destino.
1. Sector de la industria a la que pertenece la estación.


Desde la serie más atómica a la agrupación más general, el orden es el siguiente:
id_bomba --> id_equipo --> id_empresa --> id_industria

se puede acceder la URL:

https://www.kaggle.com/datasets/danielabosch/fuel-transactions-from-gas-pump/data

Los atributos del archivo son:

1. unique_id_transa:	Id único de transacción.
1. id_transaccion: Id registrado por un nodo IoT. Puede repetirse entre distintos nodos.
1. id_vehiculo: 	Id único de unidad de consumo. Puede ser un vehículo u otra unidad.
1. id_usuario: 	Id único del operador encargado de autorizar el despacho.
1. id_equipo: 	Id único de nodo IoT. Un nodo pertenece a una estación.
1. veh_site_id: Id de unidad de consumo registrado por el nodo. Puede repetirse entre distintos nodos.
1. pump_site_id: Id de surtidor registrado por el nodo. Puede repetirse entre distintos nodos.
1. ank_site_id: Id de tanque registrado por el nodo. Puede repetirse entre nodos.
1. user_site_id: Id de operador registrado por el nodo. Puede repetirse entre nodos.
1. producto: Id de producto registrado en el nodo IoT. Puede repetirse entre nodos.
1. id_bomba: Id único de surtidor.
1. id_tanque: Id único de tanque conectado a uno o más surtidores.
1. departamento: Departamento perteneciente al vehículo.
1. cantidad: Litros transaccionados.
1. codigo_error: Código de finalización de la operación (No indica necesariamente error).
1. timestamp: Timestamp de inicio de transacciñon
1. timestamp_stop: Timestamp de fin de transacciñon
1. nombre_prod: Producto despachado
1. veh_efficiency:  
1. volumen_inicial: Volumen inicial registrado en el tanque
1. volumen_final: Volumen final registrado en el tanque
1. temp_inicial: Temperatura registrada al iniciar la operación.
1. temp_final: Temperatura registrada al finalizar la operación.
1. volumen_comp_15_inicial: Volumen inicial registrado en el tanque compensados a 15ºC.
1. volumen_comp_15_final: Volumen final registrado en el tanque compensados a 15ºC.
1. cantidad_comp_15: Litros despachados compensados a 15ºC
1. codigo_producto: Código utilizado por la empresa para identificar productos entre nodos.
1. geo_latitud: Geo Latitud
1. geo_longitud: Geo Longitud
1. id_empresa: 	Id único de empresa a la que pertenece la estación
1. id_industria: Id único de sector de la industria a la que pertenece la empresa
1. industria: Nombre de la industria
