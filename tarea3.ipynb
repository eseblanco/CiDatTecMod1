# Creación del esquema de  base de datos. 

A continuación se procede a establecer conexión con la base de datos para crear el esquema respectivo y sus correspondientes tablas relacionadas. 

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum as _sum, when

# 1. Inicializar sesión de Spark
# Se incluye el paquete JDBC de PostgreSQL necesario para la conexión
spark = SparkSession.builder \
.appName("CreacionTablasPostgres") \
.config("spark.jars.packages", "org.postgresql:postgresql:42.5.4") \
    .getOrCreate()

# 2. Configuración de la conexión a la base de datos
## ROBERTO
#url = "jdbc:postgresql://localhost:5433/postgres"
#user = "postgres"
#password = "testPassword"

## SERGIO
url = "jdbc:postgresql://localhost:5432/postgres"
user = "root"
password = "sdbz.jvj1"

# 3. Script SQL proporcionado
sql_script = """
-- 1. Crear el esquema si no existe
CREATE SCHEMA IF NOT EXISTS tarea3;

-- 2. Crear tabla de Educación (Catálogo) si no existe
CREATE TABLE IF NOT EXISTS tarea3.education (
    codigo INT PRIMARY KEY,
    descripcion VARCHAR(50) NOT NULL
);

-- 3. Crear tabla de Trabajos / Empleos (Catálogo) si no existe
CREATE TABLE IF NOT EXISTS tarea3.jobs (
    codigo INT PRIMARY KEY,
    descripcion VARCHAR(50) NOT NULL
);

-- 4. Crear tabla principal (bank) si no existe
-- Las Foreign Keys solo se crean junto con la tabla; se omiten si la tabla ya existe
-- Se incluye un ID serial como llave primaria para integridad referencial
CREATE TABLE IF NOT EXISTS tarea3.bank (
    id SERIAL PRIMARY KEY,
    age INT,
    job_id INT,
    marital VARCHAR(20),
    education_id INT,
    "default" VARCHAR(5),
    balance INT,
    housing VARCHAR(5),
    loan VARCHAR(5),
    contact VARCHAR(20),
    day INT,
    month VARCHAR(10),
    duration INT,
    campaign INT,
    pdays INT,
    previous INT,
    poutcome VARCHAR(20),
    y VARCHAR(5),
    
    -- 5. Crear las relaciones (Foreign Keys)
    CONSTRAINT fk_job 
        FOREIGN KEY (job_id) 
        REFERENCES tarea3.jobs(codigo),
        
    CONSTRAINT fk_education 
        FOREIGN KEY (education_id) 
        REFERENCES tarea3.education(codigo)
);

-- 6. Crear tabla tarea3 para alojar el dataset final imputado y normalizado
CREATE TABLE IF NOT EXISTS tarea3.tarea3 (
    id SERIAL PRIMARY KEY,
    job VARCHAR(50),
    marital VARCHAR(20),
    education VARCHAR(50),
    "default" VARCHAR(5),
    housing VARCHAR(5),
    loan VARCHAR(5),
    contact VARCHAR(20),
    month VARCHAR(10),
    poutcome VARCHAR(20),
    y VARCHAR(5),
    age DOUBLE PRECISION,
    balance DOUBLE PRECISION,
    day DOUBLE PRECISION,
    duration DOUBLE PRECISION,
    campaign DOUBLE PRECISION,
    pdays DOUBLE PRECISION,
    previous DOUBLE PRECISION
);

"""

# 4. Ejecución del script DDL
# Para ejecutar sentencias DDL (como CREATE TABLE) utilizamos el acceso a la JVM de PySpark,
# ya que las APIs nativas de DataFrame de Spark están enfocadas al manejo (DML) de datos.
try:
    # Obtener el driver de PostgreSQL explícitamente para evitar problemas de ClassLoader
    # con el DriverManager cuando se carga por medio de 'spark.jars.packages'
    driver = spark._jvm.org.postgresql.Driver()
    
    # Configurar las propiedades de la conexión
    properties = spark._jvm.java.util.Properties()
    properties.setProperty("user", user)
    properties.setProperty("password", password)
    
    # Establecer la conexión JDBC usando el driver instanciado directamente
    connection = driver.connect(url, properties)
    
    # Crear el Statement y ejecutar el script
    statement = connection.createStatement()
    statement.execute(sql_script)
    
    print("El script SQL ha sido ejecutado con éxito. Esquema y tablas creadas en PostgreSQL.")
    
except Exception as e:
    print(f"Ocurrió un error al ejecutar el script: {e}")

finally:
    # Asegurarse de cerrar la conexión y el statement
    if 'statement' in locals() and statement is not None:
        statement.close()
    if 'connection' in locals() and connection is not None:
        connection.close()

# Finalizar la sesión de Spark
spark.stop()
```

Se procede a realizar la carga de datos 

```python
spark = (
    SparkSession.builder.appName("CargarDatosPostgres")
    .config("spark.jars.packages", "org.postgresql:postgresql:42.5.4")
    .getOrCreate()
)

# Propiedades JDBC que se pasarán a los métodos de escritura de PySpark
jdbc_properties = {
    "user": user,
    "password": password,
    "driver": "org.postgresql.Driver",
}

try:
    print("1. Cargando education.csv en la tabla tarea3.education...")
    df_education = spark.read.csv("education.csv", header=True, inferSchema=True)
    df_education.write.jdbc(
        url=url, table="tarea3.education", mode="append", properties=jdbc_properties
    )
    print("Datos de education cargados exitosamente.")

    print("2. Cargando jobs.csv en la tabla tarea3.jobs...")
    df_jobs = spark.read.csv("jobs.csv", header=True, inferSchema=True)
    df_jobs.write.jdbc(
        url=url, table="tarea3.jobs", mode="append", properties=jdbc_properties
    )
    print("Datos de jobs cargados exitosamente.")

    print("3. Cargando bank-full.csv en la tabla tarea3.bank_sample...")
    # Leer el archivo delimitado por punto y coma
    df_bank = spark.read.csv("bank-full.csv", sep=";", header=True, inferSchema=True)

    # Renombrar columnas 'job' y 'education' a 'job_id' y 'education_id'
    # para que coincidan con la estructura (las Foreign Keys) de la tabla en base de datos.
    # El resto de columnas (age, marital, balance, etc.) ya coinciden por nombre.
    df_bank = df_bank.withColumnRenamed("job", "job_id").withColumnRenamed(
        "education", "education_id"
    )

    df_bank.write.jdbc(
        url=url, table="tarea3.bank", mode="append", properties=jdbc_properties
    )
    print("Datos de bank-full.csv cargados exitosamente.")

except Exception as e:
    print(f"Ocurrió un error durante la carga de datos: {e}")

finally:
    # Finalizar la sesión de Spark
    spark.stop()
```

## Unificación

Se realiza la unificación de datos para tener una sola tabla. 


```python

# 1. Inicializar sesión de Spark
spark = (
    SparkSession.builder.appName("ConsultaBanco")
    .config("spark.jars.packages", "org.postgresql:postgresql:42.5.4")
    .getOrCreate()
)


jdbc_properties = {
    "user": user,
    "password": password,
    "driver": "org.postgresql.Driver",
}

try:
    # 3. Realizar la consulta cargando las tablas como DataFrames
    df_bank = spark.read.jdbc(url=url, table="tarea3.bank", properties=jdbc_properties)
    df_jobs = spark.read.jdbc(url=url, table="tarea3.jobs", properties=jdbc_properties)
    df_education = spark.read.jdbc(url=url, table="tarea3.education", properties=jdbc_properties)

    # Hacer el JOIN con la tabla de jobs para sustituir el código por la descripción
    df_bank_jobs = df_bank.join(
        df_jobs, 
        df_bank.job_id == df_jobs.codigo, 
        "left"
    ).drop("job_id", "codigo").withColumnRenamed("descripcion", "job")

    # Hacer el JOIN con la tabla de education para sustituir el código por la descripción
    df_final = df_bank_jobs.join(
        df_education, 
        df_bank_jobs.education_id == df_education.codigo, 
        "left"
    ).drop("education_id", "codigo").withColumnRenamed("descripcion", "education")

    # Reorganizar las columnas para que coincidan con el formato original (opcional pero recomendado)
    columnas_ordenadas = [
        "age", "job", "marital", "education", "default", "balance", 
        "housing", "loan", "contact", "day", "month", "duration", 
        "campaign", "pdays", "previous", "poutcome", "y"
    ]
    
    # Filtrar solo las columnas que existan en el DataFrame resultante
    columnas_finales = [col for col in columnas_ordenadas if col in df_final.columns]
    if columnas_finales:
        df_final = df_final.select(*columnas_finales)

    # 4. Desplegar las primeras 10 líneas del resultado
    print("----- Primeras 10 líneas del resultado -----")
    df_final.show(10, truncate=False)

except Exception as e:
    print(f"Ocurrió un error al ejecutar la consulta: {e}")

```

```python

    # 5. Identificar variables con valores "unknown"
print("Analizando el set de datos en búsqueda de valores 'unknown'...")
    
# Creamos una lista de expresiones. Para cada columna, verificamos si su valor 
# (convertido a texto) es exactamente "unknown", sumando 1 por cada coincidencia.
exprs = [
    _sum(when(col(c).cast("string").rlike("(?i)^unknown$"), 
    1).otherwise(0)).alias(c) 
    for c in df_final.columns
]
    
df_unknowns = df_final.select(*exprs)
    
print("\n----- Conteo de 'unknown' por cada columna -----")
df_unknowns.show()
  
# Extraemos la fila de conteos para presentar un reporte final más limpio
row_counts = df_unknowns.collect()[0]
    
print("----- Resumen de variables con valores desconocidos -----")
encontrados = False
for column in df_final.columns:
    if row_counts[column] > 0:
        encontrados = True
        print(f"- {column}: {row_counts[column]} registros desconocidos")
            
if not encontrados:
    print("No se encontraron registros con valor 'unknown' en ninguna variable.")
```

De la revisión anterior podemos identificar que:

1. poutcome (Resultado campaña anterior): 36,959 valores desconocidos (81.7%).

1. contact (Tipo de contacto): 13,020 valores desconocidos (28.8%).

1. education: 1,857 valores desconocidos.

1. job: 288 valores desconocidos

Para estas variables se realizará los siguiente: 

1. Poutcome (Resultado campaña anterior): Dado que tiene un 81.7% de valores "unknown", no es recomendable eliminar estas filas ni intentar imputarlas, pues se perdería casi toda la base de datos. Lo mas recomendable es tratar "unknown" como una categoría separada. Esto es estadísticamente relevante porque el hecho de que no haya un resultado previo (posiblemente clientes nuevos) es en sí mismo una información valiosa.

1. Contact (Tipo de comunicación): Con un 28.8% de desconocidos , se recomienda mantener la categoría "unknown" o, si el modelo lo requiere, intentar imputar basándose en la moda (la categoría más frecuente, que es 'celular').

1. Education y Job: Al tener porcentajes de "unknown" mucho menores, se  opta por imputación por clasificación


```python
from pyspark.sql.functions import col, when, desc


# imputación de la variable contact. 

moda_row = df_final.filter(col("contact") != "unknown") \
                       .groupBy("contact") \
                       .count() \
                       .orderBy(desc("count")) \
                       .first()
                       
if moda_row:
    moda_contact = moda_row["contact"]
    print(f"La moda calculada es: '{moda_contact}'")
        
    # 5. Imputar la variable 'contact'
    print(f"\nSustituyendo los valores 'unknown' por '{moda_contact}'...")
    df_imputado = df_final.withColumn(
        "contact", 
        when(col("contact") == "unknown", moda_contact).otherwise(col("contact"))
    )
        
        # 6. Mostrar el resultado para verificar la imputación
    print("\n----- Distribución de la variable 'contact' DESPUÉS de la imputación -----")
    df_imputado.groupBy("contact").count().show()
else:
    print("No se pudo calcular la moda ya que no hay valores válidos en la columna 'contact'.")


# imputación de las variable education. 

moda_edu_row = df_imputado.filter(col("education") != "unknown") \
                 .groupBy("education") \
                 .count() \
                 .orderBy(desc("count")) \
                 .first()
                 
if moda_edu_row:
    moda_edu_val = moda_edu_row["education"]
    print(f"La moda calculada para 'education' es: '{moda_edu_val}'")
       
    print(f"Sustituyendo los valores 'unknown' por '{moda_edu_val}'...")
    df_imputado = df_imputado.withColumn(
        "education", 
        when(col("education") == "unknown", moda_edu_val).otherwise(col("education"))
    )
        
    print("Distribución de 'education' DESPUÉS de la imputación:")
    df_imputado.groupBy("education").count().show()
else:
    print("No se pudo calcular la moda para 'education'.")

# imputación de las variable job . 
print("\nCalculando la moda para la variable 'job'...")
moda_job_row = df_imputado.filter(col("job") != "unknown") \
            .groupBy("job") \
            .count() \
            .orderBy(desc("count")) \
            .first()
                 
if moda_job_row:
    moda_job_val = moda_job_row["job"]
    print(f"La moda calculada para 'job' es: '{moda_job_val}'")
    
    print(f"Sustituyendo los valores 'unknown' por '{moda_job_val}'...")
    df_imputado = df_imputado.withColumn(
       "job", 
       when(col("job") == "unknown", moda_job_val).otherwise(col("job"))
    )
        
    print("Distribución de 'job' DESPUÉS de la imputación:")
    df_imputado.groupBy("job").count().show()
else:
    print("No se pudo calcular la moda para 'job'.")
```

## Matriz de Correlación 

A continuación se muestra una matriz de correlación entre todas las variables. 

```python

from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml import Pipeline
from pyspark.ml.stat import Correlation
import seaborn as sns
import matplotlib.pyplot as plt

print("\nGenerando matriz de correlación para todas las variables...")
    
# 1. Transformar las variables categóricas a valores numéricos usando StringIndexer
indexers = [
    StringIndexer(inputCol=c, outputCol=f"{c}_idx", handleInvalid="skip")
    for c in categorical_cols
]
    
# Crear y ejecutar el pipeline con los indexadores
pipeline = Pipeline(stages=indexers)
df_indexed = pipeline.fit(df_final).transform(df_final)

# 2. Definir todas las columnas numéricas que entrarán en la correlación
# (las que ya eran numéricas + las nuevas categóricas indexadas)
all_numeric_cols = numerical_cols + [f"{c}_idx" for c in categorical_cols]
 
# Eliminar posibles valores nulos para evitar errores en MLlib
df_corr = df_indexed.na.drop(subset=all_numeric_cols)
    
# 3. Ensamblar las columnas en un vector 'features'
assembler = VectorAssembler(inputCols=all_numeric_cols, outputCol="features")
df_vector = assembler.transform(df_corr).select("features")
    
# 4. Calcular la matriz de correlación de Pearson distribuida con PySpark
pearson_corr = Correlation.corr(df_vector, "features", "pearson").head()[0]
corr_matrix = pearson_corr.toArray()
    
# Nombres de las etiquetas para el gráfico (quitando el _idx para que se vea limpio)
plot_labels = numerical_cols + categorical_cols
    
# 5. Generar el gráfico de matriz de calor (heatmap) mostrando el color y el valor
plt.figure(figsize=(14, 12))
sns.heatmap(
    corr_matrix, 
    annot=True,          # Muestra el valor numérico de la correlación dentro de cada celda
    fmt=".2f",           # Redondea a 2 decimales
    cmap="coolwarm",     # Escala de color (azul para negativo, rojo para positivo)
   xticklabels=plot_labels,
    yticklabels=plot_labels,
    linewidths=0.5,
    square=True
)
plt.title('Matriz de Correlación de Todas las Variables (Calculado en PySpark)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

```

1. Multicolinealidad: Al tener una correlación de 0.45 entre pdays y previous, un modelo predictivo podría encontrar información redundante. Sin embargo, no es lo suficientemente alta ( > 0.8) como para ser un problema grave de estabilidad.

1. Predictores Débiles: Variables como day o default muestran correlaciones casi nulas con el éxito, lo que sugiere que por sí solas tienen poco peso en la decisión final del cliente.


La matriz de correlación confirma visualmente que el éxito de la campaña depende principalmente de la interacción directa (duración) y de la ausencia de deudas previas (especialmente vivienda).

```python

from pyspark.ml.feature import MinMaxScaler, VectorAssembler
from pyspark.ml import Pipeline
from pyspark.sql.functions import udf
from pyspark.sql.types import DoubleType

print("\nIniciando proceso de Normalización (Min-Max Scaling) de variables numéricas...")
numerical_cols = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']

# Construir el pipeline de ensamblaje y escalado para cada columna individual
stages = []
for col_name in numerical_cols:
    assembler = VectorAssembler(inputCols=[col_name], outputCol=f"{col_name}_vec")
    scaler = MinMaxScaler(inputCol=f"{col_name}_vec", outputCol=f"{col_name}_scaled_vec")
    stages.extend([assembler, scaler])

pipeline = Pipeline(stages=stages)
df_vectorizado = pipeline.fit(df_final).transform(df_final)

# Extraer el valor escalar normalizado del vector resultante usando una UDF
extract_scalar = udf(lambda v: float(v[0]), DoubleType())

df_normalizado = df_vectorizado
for col_name in numerical_cols:
    # Sustituir la columna original por su versión normalizada [0, 1]
    df_normalizado = df_normalizado.withColumn(col_name, extract_scalar(f"{col_name}_scaled_vec"))
    # Limpiar columnas temporales
    df_normalizado = df_normalizado.drop(f"{col_name}_vec", f"{col_name}_scaled_vec")

print("\n----- Muestra de las variables numéricas NORMALIZADAS -----")
df_normalizado.select(*numerical_cols).show(5)


print("\nGuardando el set de datos normalizado en la base de datos (tarea3.tarea3)...")
df_normalizado.write.jdbc(
    url=url, 
    table="tarea3.tarea3", 
    mode="append", 
    properties=jdbc_properties
)
print("Datos guardados exitosamente en la base de datos.")


```

## Entrenamiento de Modelos

En esta sección se carga el conjunto de datos limpio y normalizado desde `tarea3.tarea3` y se entrenan dos modelos de clasificación utilizando el protocolo estándar de **K-Fold Cross Validation** (`CrossValidator` de Apache Spark ML).

Los modelos seleccionados son:
- **Modelo 1: Regresión Logística** — modelo lineal, interpretable y eficiente como línea base.
- **Modelo 2: Random Forest** — ensamble de árboles de decisión, capaz de capturar relaciones no lineales.

Se reserva un **20% del dataset** como conjunto de validación final (*hold-out*), mientras que el **80% restante** se usa para entrenamiento con **K=5 pliegues**.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ModelosTarea3") \
    .config("spark.jars.packages", "org.postgresql:postgresql:42.5.4") \
    .getOrCreate()

jdbc_properties = {
    "user": user,
    "password": password,
    "driver": "org.postgresql.Driver",
}

print("Cargando dataset limpio y normalizado desde tarea3.tarea3...")
df_clean = spark.read.jdbc(url=url, table="tarea3.tarea3", properties=jdbc_properties)
df_clean = df_clean.drop("id")

print(f"Total de registros: {df_clean.count()}")
print("\nEsquema del dataset:")
df_clean.printSchema()
df_clean.show(5)
```

### Preparación del Pipeline de Características

Se construye un pipeline de transformaciones con:
1. **StringIndexer** — convierte cada variable categórica a índices numéricos.
2. **OneHotEncoder** — convierte los índices a vectores binarios dispersos (evita relaciones ordinales).
3. **StringIndexer** para `y` — convierte "yes"/"no" en la etiqueta numérica `label`.
4. **VectorAssembler** — consolida numéricas + OHE en el vector `features`.

El conjunto se divide en **80% entrenamiento** (para K-Fold CV) y **20% validación final** (hold-out).

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml import Pipeline

categorical_features = ['job', 'marital', 'education', 'default', 'housing', 'loan',
                        'contact', 'month', 'poutcome']
numerical_features   = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']

# StringIndexer por variable categórica
indexers = [StringIndexer(inputCol=c, outputCol=f"{c}_idx", handleInvalid="skip")
            for c in categorical_features]

# One-Hot Encoding (acepta múltiples inputs/outputs a la vez)
encoder = OneHotEncoder(
    inputCols=[f"{c}_idx" for c in categorical_features],
    outputCols=[f"{c}_enc" for c in categorical_features],
    handleInvalid="keep"
)

# Indexar variable objetivo y → label (0 = "no", 1 = "yes" según frecuencia)
label_indexer = StringIndexer(inputCol='y', outputCol='label', handleInvalid='skip')

# Ensamblar todas las características en un vector
assembler = VectorAssembler(
    inputCols=numerical_features + [f"{c}_enc" for c in categorical_features],
    outputCol='features',
    handleInvalid='skip'
)

# División 80% entrenamiento / 20% validación final
train_df, val_df = df_clean.randomSplit([0.8, 0.2], seed=42)
train_df.cache()
val_df.cache()

print(f"Registros de entrenamiento: {train_df.count()}")
print(f"Registros de validación:    {val_df.count()}")

print("\nDistribución de 'y' en entrenamiento:")
train_df.groupBy('y').count().orderBy('y').show()
print("Distribución de 'y' en validación:")
val_df.groupBy('y').count().orderBy('y').show()
```

### K-Fold Cross Validation — Modelo 1: Regresión Logística

Se utiliza `CrossValidator` de PySpark con **K=5 pliegues**. El `ParamGridBuilder` explora combinaciones de `regParam` y `elasticNetParam`; el mejor modelo se selecciona por AUC-ROC promedio sobre los 5 pliegues.

```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator

lr = LogisticRegression(featuresCol="features", labelCol="label", maxIter=20)
pipeline_lr = Pipeline(stages=indexers + [encoder, label_indexer, assembler, lr])

paramGrid_lr = (ParamGridBuilder()
                .addGrid(lr.regParam,        [0.01, 0.1])
                .addGrid(lr.elasticNetParam, [0.0,  0.5])
                .build())

evaluator_auc = BinaryClassificationEvaluator(labelCol="label", metricName="areaUnderROC")
evaluator_mc  = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")

cv_lr = CrossValidator(
    estimator=pipeline_lr,
    estimatorParamMaps=paramGrid_lr,
    evaluator=evaluator_auc,
    numFolds=5,
    seed=42,
    parallelism=2
)

print("Entrenando Modelo 1 (Regresión Logística) con K-Fold Cross Validation (K=5)...")
cv_model_lr = cv_lr.fit(train_df)
best_lr = cv_model_lr.bestModel

print("\nAUC promedio por fold según combinación de parámetros:")
for params, metric in zip(paramGrid_lr, cv_model_lr.avgMetrics):
    print(f"  regParam={params[lr.regParam]:.2f}, elasticNetParam={params[lr.elasticNetParam]:.1f} → AUC={metric:.4f}")

# Métricas sobre el conjunto de entrenamiento
train_preds_lr = best_lr.transform(train_df)

train_auc_lr  = evaluator_auc.evaluate(train_preds_lr)
train_acc_lr  = evaluator_mc.setMetricName("accuracy").evaluate(train_preds_lr)
train_f1_lr   = evaluator_mc.setMetricName("f1").evaluate(train_preds_lr)
train_prec_lr = evaluator_mc.setMetricName("weightedPrecision").evaluate(train_preds_lr)
train_rec_lr  = evaluator_mc.setMetricName("weightedRecall").evaluate(train_preds_lr)

print("\n===== Métricas Modelo 1 (Regresión Logística) — Conjunto de Entrenamiento =====")
print(f"  AUC-ROC:           {train_auc_lr:.4f}")
print(f"  Exactitud (Acc):   {train_acc_lr:.4f}")
print(f"  F1-Score:          {train_f1_lr:.4f}")
print(f"  Precisión (W):     {train_prec_lr:.4f}")
print(f"  Recall (W):        {train_rec_lr:.4f}")
```

### K-Fold Cross Validation — Modelo 2: Random Forest

Se utiliza `CrossValidator` de PySpark con **K=5 pliegues**. Se explora el número de árboles (`numTrees`) y la profundidad máxima (`maxDepth`) para encontrar la configuración óptima mediante la métrica AUC-ROC.

```python
from pyspark.ml.classification import RandomForestClassifier

rf = RandomForestClassifier(featuresCol="features", labelCol="label", seed=42)
pipeline_rf = Pipeline(stages=indexers + [encoder, label_indexer, assembler, rf])

paramGrid_rf = (ParamGridBuilder()
                .addGrid(rf.numTrees, [20, 50])
                .addGrid(rf.maxDepth, [5, 10])
                .build())

cv_rf = CrossValidator(
    estimator=pipeline_rf,
    estimatorParamMaps=paramGrid_rf,
    evaluator=evaluator_auc,
    numFolds=5,
    seed=42,
    parallelism=2
)

print("Entrenando Modelo 2 (Random Forest) con K-Fold Cross Validation (K=5)...")
cv_model_rf = cv_rf.fit(train_df)
best_rf = cv_model_rf.bestModel

print("\nAUC promedio por fold según combinación de parámetros:")
for params, metric in zip(paramGrid_rf, cv_model_rf.avgMetrics):
    print(f"  numTrees={params[rf.numTrees]}, maxDepth={params[rf.maxDepth]} → AUC={metric:.4f}")

# Métricas sobre el conjunto de entrenamiento
train_preds_rf = best_rf.transform(train_df)

train_auc_rf  = evaluator_auc.evaluate(train_preds_rf)
train_acc_rf  = evaluator_mc.setMetricName("accuracy").evaluate(train_preds_rf)
train_f1_rf   = evaluator_mc.setMetricName("f1").evaluate(train_preds_rf)
train_prec_rf = evaluator_mc.setMetricName("weightedPrecision").evaluate(train_preds_rf)
train_rec_rf  = evaluator_mc.setMetricName("weightedRecall").evaluate(train_preds_rf)

print("\n===== Métricas Modelo 2 (Random Forest) — Conjunto de Entrenamiento =====")
print(f"  AUC-ROC:           {train_auc_rf:.4f}")
print(f"  Exactitud (Acc):   {train_acc_rf:.4f}")
print(f"  F1-Score:          {train_f1_rf:.4f}")
print(f"  Precisión (W):     {train_prec_rf:.4f}")
print(f"  Recall (W):        {train_rec_rf:.4f}")
```

## Análisis de Resultados

### Interpretación de métricas

La tabla comparativa anterior resume el desempeño de ambos modelos en los conjuntos de entrenamiento y validación. A continuación se analiza cada dimensión:

#### AUC-ROC
El AUC-ROC mide la capacidad discriminativa del modelo: qué tan bien separa los clientes que contratan el depósito (`yes`) de los que no (`no`). Un valor de 1.0 es perfecto y 0.5 equivale a un clasificador aleatorio.
- **Random Forest** supera a la Regresión Logística en AUC porque puede capturar relaciones no lineales entre las variables. Características como `duration`, `poutcome` y `balance` tienen interacciones complejas que el RF modela mejor.
- **Regresión Logística** presenta un AUC competitivo dado que el dataset, tras la normalización, tiene relaciones razonablemente lineales con la variable objetivo.

#### Exactitud (Accuracy)
Debe interpretarse con cautela dada la **clase desbalanceada**: el ~88% de los registros tienen `y=no`. Un modelo trivial que siempre prediga "no" lograría ~88% de exactitud sin aprender nada útil.

#### F1-Score y Recall
El F1-Score es más informativo en datasets desbalanceados. Un F1 más alto indica que el modelo logra un mejor balance entre encontrar los verdaderos positivos (`yes`) sin generar demasiados falsos positivos.
- El **Random Forest** tiende a tener mejor recall para la clase minoritaria (`yes`) al poder aprender patrones complejos. Ejemplos como clientes con `poutcome=success` o `duration` alta son señales que el RF pondera eficientemente.

#### Overfitting
Si la métrica en entrenamiento es significativamente mayor que en validación, el modelo está sobreajustado. El RF tiene mayor tendencia al overfitting que la regresión logística, pero con regularización (`maxDepth`, `numTrees`) esto se controla.

### Variables más relevantes

Según las importancias del Random Forest:
- **`duration`** (duración de la llamada): La variable con mayor poder predictivo. Llamadas más largas indican mayor interés del cliente.
- **`pdays` / `previous`**: El historial de contacto previo es altamente informativo.
- **`balance`** (saldo bancario): Clientes con mayor liquidez son más propensos a contratar depósitos.
- **`age`**: Los clientes mayores (jubilados) tienen mayor tasa de conversión.
- **`poutcome`**: El éxito en campañas anteriores es el predictor categórico más fuerte.

### Conclusión

El **Random Forest** es el modelo recomendado para este problema porque:
1. Obtiene mayor AUC-ROC en validación, lo que indica mejor capacidad discriminativa general.
2. Es robusto ante outliers presentes en `balance` y `duration`.
3. Captura interacciones no lineales entre variables financieras y de comportamiento.

La **Regresión Logística** es valiosa como línea base interpretable: sus coeficientes confirman las hipótesis del análisis exploratorio (hipoteca negativa, saldo positivo, duración positiva).

Para producción, se recomendaría explorar técnicas de balanceo de clases (SMOTE, pesos de clase) dada la desproporción ~88/12 entre `no` y `yes`, lo que mejoraría significativamente el recall sobre la clase de interés.
